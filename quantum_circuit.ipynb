```python
# creating a bell state from |00> 
from qiskit.quantum_info import Statevector
from qiskit import QuantumCircuit, transpile
from qiskit_aer import AerSimulator

# |00> --> Bell State
bell_qc = QuantumCircuit(2) # two qubits and two classical bits
bell_qc.h(0)
bell_qc.cx(0,1)
bell_qc.measure_all()
```

```python
# state = Statevector.from_label("00")
# print(state)

# final_state = state.evolve(bell_qc)
# print(final_state)
```

```python
simulator = AerSimulator()
compiled_circuit = transpile(bell_qc, simulator)
job = simulator.run(compiled_circuit, shots=1000)
result = job.result()
counts = result.get_counts(bell_qc)
print(counts)
```
